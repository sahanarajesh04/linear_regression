# Linear Regression in Python

This notebook reads `regression_data.csv`, visualizes the relationship between years of experience and salary, fits a simple linear regression model, overlays the fitted line on the scatter plot, and evaluates model performance.

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score

X_COLUMN = "YearsExperience"
Y_COLUMN = "Salary"

dataset = pd.read_csv("regression_data.csv")
dataset.head()
```

```python
plt.figure(figsize=(8, 6))
plt.scatter(dataset[X_COLUMN], dataset[Y_COLUMN], color="red", label="Observations")
plt.title("Salary vs Years of Experience")
plt.xlabel("Years of Experience")
plt.ylabel("Salary")
plt.legend()
plt.show()
```

```python
model = LinearRegression()
X = dataset[[X_COLUMN]]
y = dataset[Y_COLUMN]
model.fit(X, y)

print(f"Intercept: {model.intercept_:.2f}")
print(f"Coefficient ({X_COLUMN}): {model.coef_[0]:.2f}")
```

```python
predictions = model.predict(X)

plt.figure(figsize=(8, 6))
plt.scatter(dataset[X_COLUMN], dataset[Y_COLUMN], color="red", label="Observations")
plt.plot(dataset[X_COLUMN], predictions, color="blue", label="Regression line")
plt.title("Salary vs Years of Experience")
plt.xlabel("Years of Experience")
plt.ylabel("Salary")
plt.legend()
plt.show()
```

```python
r2 = r2_score(y, predictions)
rmse = mean_squared_error(y, predictions) ** 0.5

print(f"R-squared: {r2:.4f}")
print(f"RMSE: {rmse:.2f}")
```
