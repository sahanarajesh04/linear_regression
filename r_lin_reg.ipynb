# Linear Regression in R

This notebook reads `regression_data.csv`, visualizes the relationship between years of experience and salary, fits a simple linear regression model, overlays the fitted line on the scatter plot, and evaluates model performance.

```r
dataset <- read.csv("regression_data.csv")
head(dataset)
```

```r
plot(dataset$YearsExperience, dataset$Salary,
     col = "red", pch = 19,
     main = "Salary vs Years of Experience",
     xlab = "Years of Experience",
     ylab = "Salary")
```

```r
model <- lm(Salary ~ YearsExperience, data = dataset)
coef(model)
```

```r
library(ggplot2)

ggplot(dataset, aes(x = YearsExperience, y = Salary)) +
  geom_point(color = "red") +
  geom_line(aes(y = predict(model, newdata = dataset)), color = "blue") +
  ggtitle("Salary vs Years of Experience") +
  xlab("Years of Experience") +
  ylab("Salary")
```
